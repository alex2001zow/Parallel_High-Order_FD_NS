```python
import numpy as np
import matplotlib.pyplot as plt

# Data points and error values
points = np.array([8, 16, 32, 64])
three_point_stencil_errors = np.array([1.7e-2, 3.6e-3, 8.6e-4, 2.1e-4])
five_points_stencil_errors = np.array([1.1e-3, 3.2e-5, 8.9e-7, 1.4e-7])

# Calculate h for reference lines
h = 1 / points

# Adjusted reference lines to start from the first stencil error point
# For 3-point stencil using h^2
reference_errors = (h**2) * (three_point_stencil_errors[0] / (h[0]**2))

# For 5-point stencil using h^4
reference_errors_h4 = (h**4) * (five_points_stencil_errors[0] / (h[0]**4))

# Plotting
plt.figure(figsize=(10, 6))
plt.loglog(points, three_point_stencil_errors, 'o-', label='3-Point stencil')
plt.loglog(points, five_points_stencil_errors, 's-', color='blue', label='5-Point stencil')
plt.loglog(points, reference_errors, 'k--', label='Reference $h^2$')
plt.loglog(points, reference_errors_h4, 'r--', label='Reference $h^4$')
plt.xlabel('Number of Points')
plt.ylabel('Error')
plt.title('Convergence of Numerical Solvers for 1D-Poisson problem')
plt.legend()
plt.grid(True)
plt.show()
```
